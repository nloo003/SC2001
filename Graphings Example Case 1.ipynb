```python
x = []
y = []
for i in range(1,10000001):
    x.append(i)
    #print(x[i-1])
    y.append(i * math.log2(i))
```

```python

    
f = plt.figure(figsize = (16,16))
ax = f.add_subplot()    
ax.plot(x, y, color = "orange")
plt.gca().yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))
plt.gca().xaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

ax.set_xlabel("Data Inputs")
ax.set_ylabel("Key Comparisons")
ax.set_title("MergeSort Time Complexity (Theoretical)")
```

```python
x = []
y = []
x1 = []
y1 = []
S = 4
for i in range(1,10000001):
    x.append(i)
    #print(x[i-1])
    y.append(i*S + i * math.log(i/S))
    x1.append(i)
    y1.append(i + i * math.log(i/S))
    

    
```

```python
f = plt.figure(figsize = (16,16))
ax = f.add_subplot()    
ax.plot(x, y, color = "red",dashes=[6, 2])
ax.plot(x1,y1,color = "blue", dashes=[6, 2])
plt.gca().yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))
plt.gca().xaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

ax.set_xlabel("Data Inputs")
ax.set_ylabel("Key Comparisons")
ax.set_title("Hybrid Time Complexity (Theoretical)")
```
